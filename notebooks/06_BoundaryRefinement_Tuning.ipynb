## Quick test on three cases

```python
from openplaque.boundary_tuning import tune_boundary_refinement, summarize_tuning, parameter_grid

small_grid = parameter_grid(
    min_component_voxels=(0, 5, 10),
    lumen_distance_voxels=(0, 1),
    erosion_iterations=(0,),
    high_hu_threshold=(None,),
    low_hu_threshold=(None,),
)

df_test = tune_boundary_refinement(
    sample_root="/content/sample_dataset/Sample_Dataset",
    pred_dir="/content/openplaque_predictions",
    max_cases=3,
    grid=small_grid,
    overwrite_predictions=False,
)

summary_test = summarize_tuning(df_test)
summary_test.head(10)
```

## Full tuning run

```python
full_grid = parameter_grid(
    min_component_voxels=(0, 5, 10, 20, 50),
    lumen_distance_voxels=(0, 1, 2),
    erosion_iterations=(0, 1),
    high_hu_threshold=(None,),
    low_hu_threshold=(None,),
)

df = tune_boundary_refinement(
    sample_root="/content/sample_dataset/Sample_Dataset",
    pred_dir="/content/openplaque_predictions",
    max_cases=None,
    grid=full_grid,
    overwrite_predictions=False,
)

summary = summarize_tuning(df)
summary.head(20)
```

```python
import os

out_dir = "/content/drive/MyDrive/OpenPlaque/Tuning"
os.makedirs(out_dir, exist_ok=True)

df.to_csv(f"{out_dir}/boundary_refinement_case_results.csv", index=False)
summary.to_csv(f"{out_dir}/boundary_refinement_parameter_summary.csv", index=False)

with open(f"{out_dir}/best_boundary_refinement.txt", "w") as f:
    f.write("Best boundary refinement parameters\n")
    f.write(str(summary.iloc[0]))
    f.write("\n")

print("Saved tuning outputs to:", out_dir)
print("\nBest parameters:")
print(summary.iloc[0])
```

```python
raw = summary[
    (summary["min_component_voxels"] == 0)
    & (summary["lumen_distance_voxels"] == 0)
    & (summary["erosion_iterations"] == 0)
].head(1)

best = summary.head(1)

print("RAW BASELINE")
print(raw.T)

print("\nBEST REFINEMENT")
print(best.T)
```
